```python
input_file_name = "input.txt"
line_index = 0
board_game_index = 0
board_game_line = 0
board_games = {}
import numpy as np

with open(input_file_name, 'r') as f:
    for line in f:
        if line_index == 0:
            first_line = line.rstrip('\n').split(",")
        else:
            if(board_game_line == 0):
                board_games[board_game_index] = line.rstrip('\n').split(" ")
                #print(line_index,":", line)
                board_game_line += 1
            else:
                board_games[board_game_index] += line.rstrip('\n').split(" ")
                #print(line_index,":", line)
                board_game_line += 1
                if(board_game_line == 6):
                    board_game_line = 0
                    board_game_index += 1
        line_index += 1
        
first_line = [int(i) for i in first_line]
# print(first_line)

for board_game in board_games:
    while('' in board_games[board_game]):
        board_games[board_game].remove('')

number_of_games = 0
for board_game in board_games:
    board_games[board_game] = np.array([int(i) for i in board_games[board_game]]).reshape(5,5)
    number_of_games += 1


    
def check_for_win():
    for board_game in board_games:
        row_sum = 0
        col_sum = 0
        total_sum = 0
        for i in range(5):
            row_sum = 0
            for j in range(5):
                row_sum += board_games[board_game][i][j]
                total_sum += board_games[board_game][i][j]
            # print("row at:",i,"sum=", row_sum,  board_games[board_game][i])
            if (row_sum == -5):
                return True, board_game
        print("total_sum now:", total_sum)
        for i in range(5):
            col_sum = 0
            for j in range(5):
                col_sum += board_games[board_game][j][i]
            # print("col at:",i,"sum=", col_sum)
            if (col_sum == -5):
                return True, board_game
    return False


# try:
#         check_win, winning_person = check_for_win()
#         if check_win == True and number_of_games > 0:
#             print("winning person", winning_person)
#             del board_games[winning_person]
#             number_of_games -= 1
#         else:
#             break

moves_played = 0
for number_called in first_line:
    print("number being called: ", number_called)
    moves_played += 1
    try:
        check_win, winning_person = check_for_win()
        if check_win == True and number_of_games > 0:
            print("Found a winner player", winning_person, "after playing", moves_played)
            print(board_games[winning_person])
            last_winner = board_games[winning_person]
            del board_games[winning_person]
            number_of_games -= 1
    except:
        print("no winners")
    for board_game in board_games:
        for current_line in board_games[board_game]:
            for index, value in enumerate(current_line):
                if current_line[index] == number_called:
                    current_line[index] = -1
```

```python
def readInput4(filename):
    with open(filename) as f:
        l = f.read().split('\n\n')
        numbers = [ int(n) for n in l[0].split(',') ]
        boards = [ [ [ int(n) for n in r.split() ] for r in l[i].strip('\n').split('\n') ] for i in range(1,len(l)) ]
        return numbers, boards
```

```python
import numpy as np

def checkWinner(b):
    for r in b:
        if sum(r)==-len(r):
            return True
        
def part2(numbers,boards):
    nwin = -1
    bwin = []
    for n in numbers:
        # draw numbers an all boards
        for b in boards:
            for r in b:
                for i in range(len(r)):
                    if r[i]==n: r[i]=-1
        # check if there's a winner board, if so remove
        for b in boards:
            if checkWinner(b) or checkWinner(np.rot90(np.array(np.array(b)))) :
                if len(boards)==1:
                    unmarked = sum([ sum([ i if i!=-1 else 0 for i in r ]) for r in b ])
                    print(unmarked, n)
                    return unmarked*n
                else:
                    boards.remove(b)
                    
numbers0,boards0 = readInput4("input2.txt")
print("Test 2:",part2(numbers0,boards0))

numbers,boards = readInput4("input.txt")
print("Part 2:",part2(numbers,boards))
```
