```python
input_file_name = "input.txt"
line_index = 0
board_game_index = 0
board_game_line = 0
board_games = {}
import numpy as np

with open(input_file_name, 'r') as f:
    for line in f:
        if line_index == 0:
            first_line = line.rstrip('\n').split(",")
        else:
            if(board_game_line == 0):
                board_games[board_game_index] = line.rstrip('\n').split(" ")
                print(line_index,":", line)
                board_game_line += 1
            else:
                board_games[board_game_index] += line.rstrip('\n').split(" ")
                print(line_index,":", line)
                board_game_line += 1
                if(board_game_line == 6):
                    board_game_line = 0
                    board_game_index += 1
        line_index += 1
```

```python
def check_for_win():
    for board_game in board_games:
        row_sum = 0
        col_sum = 0
        for i in range(5):
            row_sum = 0
            for j in range(5):
                row_sum += board_games[board_game][i][j]
            # print("row at:",i,"sum=", row_sum,  board_games[board_game][i])
            if (row_sum == -5):
                return True, board_game
        for i in range(5):
            col_sum = 0
            for j in range(5):
                col_sum += board_games[board_game][j][i]
            # print("col at:",i,"sum=", col_sum)
            if (col_sum == -5):
                return True, board_game
    return False
```
